#### Matplotlib

- Basic plotting package 
- Very customisable - every single line/pixel 
- Need alot of code to write a simple plot

```python
# Make x & y variables using numpy 
x = np.arange(0, 5, 0.1)
y = np.sin(x)

print('This is x:')
print(x)

print()

print('This is y:')
print(y)
```

### Matplot lib object orientated interface

```python

fig = plt.figure()   # create figure object - contain axes, graphics, text
ax = plt.axes()      # create axes - a single pane

ax.plot(x, y) # create continuous line plot

ax.set_xlim(0,10)
ax.set_ylim(-1.5,1.5)
ax.set_title('title of plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

### Matplot lib Pyplot interface

```python
plt.plot(x, y)
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
plt.title('title of plt')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()
```

#### Scatterplot

```python
tips = sns.load_dataset("tips")
display(tips)

# set color to Yes/No depending on value in 'smoker column' 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            data=tips)

# as above plus shape of points for time column
sns.relplot(x="total_bill",
            y="tip",
            hue="smoker",
            style="time",
            data=tips);
```

```python

# col = Time -> create 2 plots (facets) split by time column 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            col="time", 
            data=tips)
```

#### Pairs plot

```python
iris = sns.load_dataset("iris")
display(iris.head())

# plot pairs plot
# assign plot to object 'g' - sets up grid
g = sns.PairGrid(iris, hue="species")

# plot histogram on diagonal
g.map_diag(plt.hist) 
# plot scatter plots in the offdiagonals
g.map_offdiag(plt.scatter)
# add legend to plot
g.add_legend();
```

### Lineplot example

```python
fmri = sns.load_dataset("fmri")
display(fmri.head())
sns.relplot(x="timepoint",
            y="signal", 
            hue="event", 
            kind="line", 
            data=fmri)

sns.relplot(x="timepoint", 
            y="signal", 
            hue="region", 
            style="event",
            dashes=False, 
            markers=True, 
            kind="line", 
            data=fmri);
```

### Facets

```python
sns.relplot(x="timepoint",
            y="signal",
            hue="subject",
            col="region",
            row="event",
            height=3,
            kind="line",
            estimator=None,
            data=fmri)
```

### Plot Categorical Data

```python
# create violin plots
g = sns.catplot(x="day", 
                y="total_bill", 
                kind="violin",
                inner=None,
                data=tips)

# add individual datapoints -> provide ax argument
sns.swarmplot(x="day",
              y="total_bill",
              color="k",
              size=3,
              data=tips,
              ax=g.ax) 
```

Rename columns and remove unwanted columns (Chr, Start, End, Strand, Length) 


```python
counts=df.drop(['Chr','Start','End','Strand','Length'],axis=1)
counts.head(10)
# This is to replace the 'sam/' and '.bam' to empty
counts.columns=counts.columns.str.replace('sam/','').str.replace('.bam','')
counts.head(10)
# This is doing the df.rename way
counts.rename(columns={'ERR1755082':'Cd4_r1','ERR1755083':'Cd4_r2','ERR1755084':'Cd4_r3','ERR1755085':'Cd8_r1','ERR1755086':'Cd8_r2','ERR1755087':'Cd8_r3'},inplace=True)
counts.head(10)
```

```python
# This is another way to delete the columns
df.drop(columns=['Chr','Start','End','Strand','Length']).head(10)
```

```python
#Set the index to 'Geneid'
counts=counts.set_index(['Geneid'])
```

Filter out genes with no counts


```python
#Create a new column to sum the counts in each row
counts['gene_count']=counts.sum(axis=1)
counts.head(10)
#Delete rows with 'gene_count=0', which also means select rows with 'gene_count>0', and delete the 'gene_count' column
filtered_counts=counts[counts.gene_count>0].drop(columns='gene_count')
filtered_counts.head(10)
```

Plot scatter plot of correlation of replicates (pairs plot) (you will need to google how to show plots in jupyter notebook)


```python
#Plot scatter plot
#amplify the original data by adding 1 and then log the orignal counts
g = sns.PairGrid(np.log2(filtered_counts +1))

# plot histogram on diagonal
g.map_diag(plt.hist,bins=100) 
# plot scatter plots in the offdiagonals
g.map_offdiag(plt.scatter)
# add legend to plot
g.add_legend()
```

Normalise read counts to total reads per sample & convert to 'counts per million (CPM)'
    - to calculte the total reads per sample and then devided by the gene numbers


```python
#Sum the total counts per sam
sum_counts=filtered_counts_gene_index.sum()
display(sum_counts)
type(sum_counts)
#Devide counts in each cell by the total counts and then scale up to counts per million
filtered_counts_CPM = filtered_counts_gene_index.div(sum_counts,axis=1)*1e6
display(filtered_counts_CPM)
```

Log transform normalised read counts


```python
log2_filtered_counts_CPM=np.log2(filtered_counts_CPM+1)
display(log2_filtered_counts_CPM)
```

Create new dataframe of 100 genes with highest average variance in log cpm across conditions


```python
#Create a new column to have the variance information for each gene
log2_filtered_counts_CPM['variance']=log2_filtered_counts_CPM.var(axis=1)
display(log2_filtered_counts_CPM)
#Sort based on variance and extract the top 100
top_100_variance=log2_filtered_counts_CPM.sort_values(by='variance',ascending=False).head(100).drop(columns='variance')
top_100_variance
```

Plot the of expression of the top 5 most variable genes (log cpm) across all samples (line plot)


```python
#Rest the Geneid to colum
top_100_variance.reset_index(inplace=True)
display(top_100_variance)
```
